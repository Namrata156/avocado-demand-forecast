```python
# get rows where none of the bag types are missing
bag_ratios = df.dropna(subset=['SmallBags','LargeBags','XLargeBags']).copy()

# calculate ratios of each bag type to TotalBags
bag_ratios['SmallRatio'] = bag_ratios['SmallBags']/bag_ratios['TotalBags']
bag_ratios['LargeRatio'] = bag_ratios['LargeBags']/bag_ratios['TotalBags']
bag_ratios['XLargeRatio'] = bag_ratios['XLargeBags']/bag_ratios['TotalBags']

# compute mean ratio for each bag size
mean_ratios = {
    'SmallRatio': bag_ratios['SmallRatio'].mean(),
    'LargeRatio': bag_ratios['LargeRatio'].mean(),
    'XLargeRatio': bag_ratios['XLargeRatio'].mean()
}

# define a function to impute missing bag values

def impute_bag_sizes(row):
    if pd.isna(row['SmallBags']) and pd.isna(row['LargeBags']) and pd.isna(row['XLargeBags']):
        row['SmallBags'] = row['TotalBags'] * mean_ratios['SmallRatio']
        row['LargeBags'] = row['TotalBags'] * mean_ratios['LargeRatio']
        row['XLargeBags'] = row['TotalBags'] * mean_ratios['XLargeRatio']
    return row

# apply the imputation function row-wise
df = df.apply(impute_bag_sizes,axis=1)
```

### Feature Engineering

```python
# Convert all columns to lowercase
df.columns = df.columns.str.lower()

# Rename volume column for easier use
df = df.rename(columns={'totalvolume': 'volume_sold'})

# Filter for TotalUS and copy it to preserve original
df_totalus = df[df['region'] == 'TotalUS'].copy().sort_values(by='date')

# Create date-related features
df_totalus['week_number'] = df_totalus['date'].dt.isocalendar().week
df_totalus['month'] = df_totalus['date'].dt.month
df_totalus['year'] = df_totalus['date'].dt.year

# Lag features
df_totalus['lag_1'] = df_totalus['volume_sold'].shift(1)
df_totalus['lag_2'] = df_totalus['volume_sold'].shift(2)
df_totalus['lag_4'] = df_totalus['volume_sold'].shift(4)

# Rolling mean (4-week)
df_totalus['rolling_mean_4'] = df_totalus['volume_sold'].rolling(window=4).mean()

# Holiday indicator
from pandas.tseries.holiday import USFederalHolidayCalendar
calendar = USFederalHolidayCalendar()

start_date = pd.to_datetime(df_totalus['date'].min())
end_date = pd.to_datetime(df_totalus['date'].max())

# Only proceed if both dates are valid
if pd.notna(start_date) and pd.notna(end_date):
    holidays = calendar.holidays(start=start_date, end=end_date)
    df_totalus['is_holiday_week'] = df_totalus['date'].isin(holidays).astype(int)
else:
    df_totalus['is_holiday_week'] = 0  # fallback

df_totalus['is_holiday_week'] = df_totalus['date'].isin(holidays).astype(int)

# Drop NA rows from lag and rolling
df_totalus = df_totalus.dropna().reset_index(drop=True)

df_totalus.head()
```

### Prophet Model

##### Prepare data for Prophet

```python
df_prophet = df_totalus[['date', 'volume_sold']].rename(columns={
    'date': 'ds',
    'volume_sold': 'y'
})
```

```python
df_prophet.head()
```

##### Fit the model

```python
from prophet import Prophet

# Initialize and fit
model = Prophet()
model.fit(df_prophet)
```

##### Make future predictions

```python
future = model.make_future_dataframe(periods=52, freq='W')
forecast = model.predict(future)
```

##### Visualize forecast

```python
fig = model.plot(forecast)
```

```python
# Merge forecast with actuals
forecast_full = df_prophet.merge(forecast[['ds', 'yhat']], on='ds', how='left')

# Plot actual vs predicted
plt.figure(figsize=(12,6))
plt.plot(forecast_full['ds'], forecast_full['y'], label='Actual')
plt.plot(forecast_full['ds'], forecast_full['yhat'], label='Forecast')
plt.legend()
plt.title('Avocado Volume Sold: Actual vs Forecast')
plt.xlabel('Date')
plt.ylabel('Volume Sold')
plt.tight_layout()
plt.show()
```
