```python
# get rows where none of the bag types are missing
bag_ratios = df.dropna(subset=['SmallBags','LargeBags','XLargeBags']).copy()

# calculate ratios of each bag type to TotalBags
bag_ratios['SmallRatio'] = bag_ratios['SmallBags']/bag_ratios['TotalBags']
bag_ratios['LargeRatio'] = bag_ratios['LargeBags']/bag_ratios['TotalBags']
bag_ratios['XLargeRatio'] = bag_ratios['XLargeBags']/bag_ratios['TotalBags']

# compute mean ratio for each bag size
mean_ratios = {
    'SmallRatio': bag_ratios['SmallRatio'].mean(),
    'LargeRatio': bag_ratios['LargeRatio'].mean(),
    'XLargeRatio': bag_ratios['XLargeRatio'].mean()
}

# define a function to impute missing bag values

def impute_bag_sizes(row):
    if pd.isna(row['SmallBags']) and pd.isna(row['LargeBags']) and pd.isna(row['XLargeBags']):
        row['SmallBags'] = row['TotalBags'] * mean_ratios['SmallRatio']
        row['LargeBags'] = row['TotalBags'] * mean_ratios['LargeRatio']
        row['XLargeBags'] = row['TotalBags'] * mean_ratios['XLargeRatio']
    return row

# apply the imputation function row-wise
df = df.apply(impute_bag_sizes,axis=1)
```

```python
df.isna().sum()
```

##### Total Volume Sold Over Time (2015-2023)

```python
plt.figure(figsize=(8,4))
df.groupby('Date')['TotalVolume'].sum().plot()
plt.title("Total Avocado Volume Sold Over Time")
plt.xlabel("Date")
plt.ylabel("Total Volume (units)")
plt.tight_layout()
plt.show()
```

```python
df['Year_Quarter'] = df['Date'].dt.to_period('Q').astype(str)
quarterly_volume = df.groupby('Year_Quarter')['TotalVolume'].sum()
```

```python
plt.figure(figsize=(12,6))
plt.plot(quarterly_volume,marker='o')
plt.title("Total Avocado Volume Sold per Quarter")
plt.xlabel("Quarter")
plt.ylabel("Total Volume")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##### Total Volume by Region

```python
df_regions = df[df['region'] != 'TotalUS']
df_us = df[df['region'] == 'TotalUS']
region_totals = df_regions.groupby('region')['TotalVolume'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,4))
region_totals.plot(kind='bar')
plt.title("Total Volume by Region")
plt.xlabel("Region")
plt.ylabel("Total Volume")
```

##### Boxplot: Price vs Volume

#####  Scatterplot: AveragePrice vs TotalVolume

```python
# Aggregate total volume sold by type
type_totals = df.groupby('type')['TotalVolume'].sum()

# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(type_totals, labels=type_totals.index, autopct='%1.1f%%', startangle=140)
plt.title("Share of Avocado Sales by Type")
plt.axis('equal')  # Equal aspect ratio ensures it's a circle
plt.show()
```

```python
sns.scatterplot(data=df, x='AveragePrice', y='TotalVolume', hue='type')
```
